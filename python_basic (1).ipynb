6. Why are lists mutable, but tuples are immutable?
- Lists are Mutable:-Lists are meant to store sequences of data that can change elements can be added, removed, or updated.
Because of this flexible use, they are implemented to allow in-place modification.
- Tuples are immutable: Tuples are meant to store fixed collections of data. Once created, their elements cannot be changed, added, or removed. This immutability helps ensure data integrity and allows tuples to be used in situations where a constant, hashable sequence is needed.

7. What is the difference between “==” and “is” operators in Python?

- == (Equality Operator)

Checks if the values of two objects are equal.
It compares the contents or data inside the objects.
Can be overridden by classes to define custom equality logic.

Example:

a = [1, 2, 3]

b = [1, 2, 3]

print(a == b)   # True, because the lists have the same contents

- is (Identity Operator):Checks if two variables refer to the exact same object in memory.It compares object identities, not their contents.Two different objects with the same value will return False.

Example:

a = [1, 2, 3]

b = [1, 2, 3]

print(a is b)   # False, because they are different objects in memory

c = a
print(a is c)   # True, because both refer to the same object

10.  What is the difference between implicit and explicit type casting
-  Implicit Type Casting (Type Coercion)

Automatically done by Python without your intervention.

Python converts one data type to another automatically to avoid errors in operations involving mixed types.

Usually happens when you mix types in expressions (e.g., int + float).

No code needed from the programmer to trigger it.

Example:

a = 5          # int

b = 3.2        # float

c = a + b      # a is implicitly converted to float

print(c)       # Output: 8.2 (float)

- Explicit Type Casting

Done manually by the programmer using built-in functions.

You explicitly convert a value from one type to another using functions like int(), float(), str(), etc.

Useful when you want precise control over data types.

Example:

a = "123"          # string

b = int(a)         # explicitly cast to int

print(b + 5)       # Output: 128 (integer addition)

13. What is the difference between for and while loops?

-  for loop

Used to iterate over a sequence (like a list, tuple, string, or range).

Runs a block of code a fixed number of times, once for each item in the sequence.

    Syntax example:

    for item in [1, 2, 3]:
       print(item)

- while loop

Repeats a block of code as long as a condition is True.

The number of iterations is not fixed and depends on when the condition becomes false.

    Syntax example:
    count = 0
    while count < 3:
      print(count)
      count += 1
